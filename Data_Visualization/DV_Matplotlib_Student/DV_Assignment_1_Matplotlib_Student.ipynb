## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.figure(figsize = (5,3))
#E=array([0.0e+00, 9.0e+16, 1.8e+17, 2.7e+17, 3.6e+17, 4.5e+17, 5.4e+17,6.3e+17, 7.2e+17, 8.1e+17, 9.0e+17])
x = np.linspace(0.1, 10, 11)
y = np.log2(x)

plt.plot(E,y,color="green",lw=2)#linewidth= de olur
plt.xlabel('M in g',fontweight='bold',fontsize=7)
plt.ylabel('E in j',fontweight='bold',fontsize=7)
plt.title('E = mc^2',fontweight='bold',fontsize=7)

```

```python
E = np.array([0.0e+00, 9.0e+16, 1.8e+17, 2.7e+17, 3.6e+17, 4.5e+17, 5.4e+17, 6.3e+17, 7.2e+17, 8.1e+17, 9.0e+17])
x = np.linspace(0.1, 10, 12)[1:]
y = np.log2(x)

plt.figure(figsize=(5, 3))

plt.plot(E, y, color="green", lw=2)
plt.xlabel('M in g', fontweight='bold', fontsize=7)
plt.ylabel('E in j', fontweight='bold', fontsize=7)
plt.title('E = mc^2', fontweight='bold', fontsize=7)

plt.ylim(0, None) 
```

```python
plt.plot(m, E, color="green", lw=3)  # Enerji-madde grafiği
plt.title("E = mc^2")                # Grafik başlığı
plt.xlabel("M in g")                 # x-ekseni etiketi
plt.ylabel("E in J")                 # y-ekseni etiketi
plt.xlim(0, 10)                      # x-ekseni sınırları
plt.yscale("log")                    # y-eksenini logaritmik ölçeğe ayarlama
plt.grid(which="both", axis="y")     # y-ekseninde ızgara çizgilerini ayarlama
plt.show()
```

```python
plt.figure(figsize=(3.5, 3))
plt.plot(m, E, color="green", lw=2.5)  # Enerji-madde grafiği
plt.title("E = mc^2")                # Grafik başlığı
plt.xlabel("M in g")                 # x-ekseni etiketi
plt.ylabel("E in J")                 # y-ekseni etiketi
plt.xlim(0, 10)                      # x-ekseni sınırları
plt.yscale("log")                    # y-eksenini logaritmik ölçeğe ayarlama
plt.grid(which="both", axis="y")     # y-ekseninde ızgara çizgilerini ayarlama
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
fig, ax1 = plt.subplots(figsize=(9, 6))

# İlk çizgi (sol y ekseninde)
ax1.set_xlabel("july16_2007")
ax1.set_ylabel("july16_2007", color="blue")
ax1.plot(labels, july16_2007, color="blue", label="july16_2007")
ax1.tick_params(axis="y", labelcolor="blue")

# İkinci çizgi (sağ y ekseninde)
ax2 = ax1.twinx()
ax2.set_ylabel('july16_2020', color="red")
ax2.plot(labels, july16_2020, color="red", label="july16_2020")
ax2.tick_params(axis="y", labelcolor="red")

plt.title(label = "___ALKIŞI ALIRIM KOLAYDI___")
```
