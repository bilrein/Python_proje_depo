## 1a: Box plot of age in both genders separeted by manner of death.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="gender", y="age", hue="manner_of_death", data=df)
plt.title("Box Plot of Age by Gender and Manner of Death")
plt.show()
```

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(14, 8))
sns.swarmplot(x="gender", y="age", hue="manner_of_death", data=df)
plt.title("Swarm Plot of Age by Gender and Manner of Death")
plt.show()
```

## 2: Plot 12 most dangerous cities

```python
most_dangerous_12 = df["city"].value_counts().head(12)
most_dangerous_12
```

```python
most_dangerous_12 =np.array(most_dangerous_12)
most_dangerous_12
```

```python
top_cities = df["city"].value_counts().head(12).index
top_cities
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=top_cities,y=most_dangerous_12,hue=top_cities)
plt.title("Most Dangerous CITY",c="red")
```

```python
# with seaborn
```

```python
df["city"].value_counts().head(12).reset_index()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=top_cities,y=most_dangerous_12,hue=top_cities)
plt.title("Most Dangerous CITY",c="blue")
plt.xticks(rotation=45)
plt.legend()
```

```python
top_cities
```

```python
plt.figure(figsize=(6,3))
top_citys = df["city"].value_counts().head(12).index
sns.countplot(x="city", data=df[df["city"].isin(top_citys)],hue="city",order=(top_cities))
plt.xticks(rotation=45)
```
